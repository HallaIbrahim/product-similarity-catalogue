### Data Extraction :

```python
# Function to extract the 'value' from a list
def extract_value(json_list):
    try:
        if isinstance(json_list, list):
            # Assuming you want to extract the 'value' from the first dictionary in the list
            return json_list[0].get('value', None)
        else:
            return None
    except (IndexError, AttributeError):
        return None

selected_columns = ['brand', 'color', 'item_name', 'model_name', 'model_number', 'model_year',
                    'product_type', 'style', 'item_weight', 'material',
                    'fabric_type', 'product_description','pattern','finish_type','item_shape']

data_frames = []

for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        with jsonlines.open(file_path) as json_file:
            json_objects = list(json_file)

            if len(json_objects) > 0:
                dff = pd.DataFrame(json_objects)

                # Apply the modified extract_value function to each cell in selected columns
                for column in selected_columns:
                    if column in dff.columns:
                        dff[column] = dff[column].apply(extract_value)

                data_frames.append(dff)

# Concatenate all data frames into a single data frame
combined_df = pd.concat(data_frames, ignore_index=True)

# Save the combined DataFrame to a CSV file
#combined_df.to_csv('outputt.csv', index=False)
```

### Data Modelling

### Recommend 5 products above 90% similarity

```python
# Convert your dataset to a list of dictionaries
products = new_data.to_dict(orient='records')

# Extract the titles for TF-IDF vectorization
titles = [product['item_name'] for product in products]

# Vectorize the titles using TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(titles)

# Train Nearest Neighbors model
knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=5)
knn_model.fit(tfidf_matrix)

# Function to recommend similar products for a given query title
def recommend_similar_products(query_title, knn_model, tfidf_vectorizer, titles, data, threshold=0.90, k=5):
    # Transform the query title into a TF-IDF vector
    query_vector = tfidf_vectorizer.transform([query_title])

    # Find the indices and distances of k-nearest neighbors
    distances, indices = knn_model.kneighbors(query_vector)

    # Filter products based on the similarity threshold
    high_similarity_indices = [i for i, distance in zip(indices.flatten(), distances.flatten()) if 1.0 - distance > threshold]

    if not high_similarity_indices:
        return "No perfect match found."

    # Return the top-k most similar products with their attribute values
    similar_products = [
        {'item_name': titles[i], 'attributes': data.iloc[i].to_dict(), 'similarity': 1.0 - distances.flatten()[j]} 
        for j, i in enumerate(high_similarity_indices)
    ]
    return similar_products

```

```python
query_title = input("Enter the item name: ")
recommended_products = recommend_similar_products(query_title, knn_model, tfidf_vectorizer, titles, new_data, threshold=0.90, k=5)
if isinstance(recommended_products, str):
    print(recommended_products)
else:
    for product in recommended_products:
        print(f"Item Name: {product['item_name']}, Similarity: {product['similarity']}")
        print("Attributes:")
        for key, value in product['attributes'].items():
            print(f"  {key}: {value}")
        print("\n")
        print("-" * 100)
        print("\n")
```

### Recommend the top product whose highest similarity score

```python
def recommend_top1_highest_similarity(query_title, knn_model, tfidf_vectorizer, titles, data):
    # Transform the query title into a TF-IDF vector
    query_vector = tfidf_vectorizer.transform([query_title])

    # Find the index and distance of the nearest neighbor
    distance, index = knn_model.kneighbors(query_vector, n_neighbors=1)

    # Get the index of the product with the highest similarity
    highest_similarity_index = index.flatten()[0]

    # Return the most similar product with its attribute values
    similar_product = {
        'item_name': titles[highest_similarity_index],
        'attributes': data.iloc[highest_similarity_index].to_dict(),
        'similarity': 1.0 - distance.flatten()[0]
    }
    return similar_product
```

```python
# Example usage:
query_title = input("Enter the item name: ")
recommended_product = recommend_top1_highest_similarity(query_title, knn_model, tfidf_vectorizer, titles, new_data)

print(f"Item Name: {recommended_product['item_name']}, Similarity: {recommended_product['similarity']}")
print("Attributes:")
for key, value in recommended_product['attributes'].items():
    print(f"  {key}: {value}")
```

### calculate the similarity score for the input product

```python
def calculate_similarity_for_new_product(new_product_title, knn_model, tfidf_vectorizer, titles, data):
    # Transform the new product title into a TF-IDF vector
    new_product_vector = tfidf_vectorizer.transform([new_product_title])

    # Find the index and distance of the nearest neighbor
    distance, index = knn_model.kneighbors(new_product_vector, n_neighbors=1)

    # Get the index of the product with the highest similarity
    highest_similarity_index = index.flatten()[0]

    # Return the most similar product with its attribute values
    similar_product = {
        'item_name': titles[highest_similarity_index],
        'attributes': data.iloc[highest_similarity_index].to_dict(),
        'similarity': 1.0 - distance.flatten()[0]
    }
    return similar_product

# Get user input for a new product
new_product_title = input("Enter the title of the new product: ")

# Calculate similarity for the new product
similar_product = calculate_similarity_for_new_product(new_product_title, knn_model, tfidf_vectorizer, titles, new_data)

# Display the result
print(f"\nMost Similar Product:")
print(f"Item Name: {similar_product['item_name']}")
print(f"Similarity: {similar_product['similarity']}")
print("Attributes:")
for key, value in similar_product['attributes'].items():
    print(f"  {key}: {value}")
```
